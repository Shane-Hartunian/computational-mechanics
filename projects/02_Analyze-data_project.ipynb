## 1. Explore data

Here, I load the data into a Pandas dataframe to see what headings and values are available. I see two columns that I want to analyze
- 'date'
- 'open'

````python
data = pd.read_csv('../data/nyse-data.csv')
data['date'] = pd.to_datetime(data['date'])
#^^^pd.to_datetime makes dates into an object that can be compared easily
data
````

I only want the `symbol == GOOGL` data, so I use a Pandas call. I also want to remove the big drop in price after Mar, 2014, so I specify the date < 2014-03-01.

````python
google_data = data[data['symbol'] == 'GOOGL']
google_date = pd.to_datetime(google_data['date'])

plt.plot(google_data['date'], google_data['open'])

# remove data > 2014-03-01
google_data_pre_2014 = google_data[google_date < pd.to_datetime('2014-03-01')]

plt.plot(google_data_pre_2014['date'], google_data_pre_2014['open'])
plt.xlabel('date')
plt.ylabel('opening price (\$)')
plt.show()
````

## 2. Data analysis

The GOOGL stock nearly doubled in price from 2010 through 2014. Day-to-day, the price fluctuates randomly. Here, I look at the fluctuations in price using [`np.diff`](https://numpy.org/doc/1.20/reference/generated/numpy.diff.html).

````python
dprice = np.diff(google_data_pre_2014['open'])  
#^^^np.diff finds the difference between each day and the next

plt.plot(google_data_pre_2014['date'][1:], dprice)
plt.xticks(rotation=45)
plt.xlabel('date')
plt.ylabel('change in opening price (\$/day)')
plt.show()
````

Looking at the price day-to-day, it would appear to be an average change of $0/day. Next, I explore the statistical results of the change in opening price

- mean
- standard deviation
- histogram

````python
mean_dprice = np.mean(dprice)
print(mean_dprice)

std_dprice = np.std(dprice)
print(std_dprice)

x = np.linspace(-40, 40)
from scipy import stats
price_pdf = stats.norm.pdf(x, loc = mean_dprice, scale = std_dprice)
#^^^loc = mean of the data, scale = std of the data
#^^^norm.pdf sets the data as a probability density function
````

````python
plt.hist(dprice, bins=50, density=True)
#^^^plots the price difference as a histogram
#^^^when density=True, returns a probability density so that the area under
#^^^integrates to 1 centered at 0
plt.plot(x, price_pdf)
plt.title('GOOGL changes in price over 4 years\n'+
         'avg: \${:.2f} stdev: \${:.2f}'.format(mean_dprice, std_dprice))
plt.show()
````

From this statistical result, it looks like the price changes followed a normal distribution with an average change of $\$0.57$ and a standard deviation of $\$9.84$.

## 3. Create random variables

Now, I know the distribution shape and characteristics to simulate the random walk price changes for the GOOGL prices each day. Here, I generate random variables with the following array structure:

Date |model 1| model 2| model 3|...| model N|
---|---|---|---|---|---|
day 1|$\Delta \$ model~1$|$\Delta \$ model~2$|$\Delta \$ model~3$|...|$\Delta \$ model~N$|
day 2|$\Delta \$ model~1$|$\Delta \$ model~2$|$\Delta \$ model~3$|...|$\Delta \$ model~N$|
...|...|...|...|...|...|

Each column is one random walk model. Each row is one simulated day. If I want to look at _one_ model predition, I would plot one column. If I want to look at the _average_ result, I take the average of each row. To start, I'll create 100 random walk models. I use the [`normal`](https://numpy.org/doc/stable/reference/random/generated/numpy.random.Generator.normal.html#numpy.random.Generator.normal) distribution to match the statistical distribution I found in part 2.

````python
rng = default_rng(42)
#^^^fixes the random number seed

N_models = 100
dprice_model = rng.normal(size = (len(google_data_pre_2014), N_models), 
                          loc = 0.568, scale = 9.838)
#^^^loc=0.568=mean_dprice; scale=9.838=std_dprice

print(len(dprice_model[0,:]))
#^^^dprice_model has different 100 random models - 100 columns - for each day

print(len(dprice_model[:,0]))
#^^^dprice_model has 1046 simulated days - 1046 rows - for each model
#^^^matches the length of google_data_pre_2014

plt.hist(dprice, 50, density=True, label = 'NYSE data')
plt.plot(x, price_pdf)
plt.hist(dprice_model[:, 0], 50, density = True, 
         histtype = 'step', linewidth = 3, 
         label = 'model prediction 1')
#^^^histtype genrates a line plot unfilled default is 'bar'

plt.title('GOOGL changes in price over 4 years\n'+
         'avg: \${:.2f} stdev: \${:.2f}'.format(mean_dprice, std_dprice))
plt.legend()
plt.show()
````

## 4. Generate random walk predictions

Above, I show that the simulated data follows the requested normal distribution. Now, I can cumulatively sum these steps to predict the stock prices each day. I use the  [`np.cumsum`](https://numpy.org/doc/1.20/reference/generated/numpy.cumsum.html) argument, `axis = 0` to sum along the columns i.e. each row becomes the sum of previous rows. 

```python
>>> a = np.array([[1,2,3], [4,5,6]])
>>> a
array([[1, 2, 3],
       [4, 5, 6]])
>>>> np.cumsum(a,axis=0)      # sum over rows for each of the 3 columns
array([[1, 2, 3],
       [5, 7, 9]])
```

Then, I plot all of the random walk models to compare to the NYSE data. The models are given transparency using the `alpha = 0.3` command (_`alpha = 0` is invisible, `alpha = 1` is opaque_).

````python
price_model = np.cumsum(dprice_model, 
                        axis = 0) + google_data_pre_2014['open'].values[0]
#print(google_data_pre_2014['open'].values[0])

plt.plot(google_data_pre_2014['date'], price_model, alpha = 0.3)

plt.plot(google_data_pre_2014['date'], google_data_pre_2014['open'], c = 'k',
         label = 'NYSE data')
plt.xlabel('date')
plt.ylabel('opening price (\$)')
plt.legend()
plt.show()
````

As you would expect, there are a wide variety of predictions for the price of GOOGL stocks using random numbers. Next, I try to get some insight into the average changes in the random walk model. I use the `np.mean` and `np.std` across the columns of the `price_model` prediction data, using `axis = 1` now.

````python
price_model_avg = np.mean(price_model, axis = 1)
price_model_std = np.std(price_model, axis = 1)
#^^^get the mean and std across the 100 columns ie for each model simulation
#^^^the mean and std is recorded for each day- length of google_data_pre_2014
print(len(price_model_avg))

plt.plot(google_data_pre_2014['date'], price_model, alpha = 0.3)

plt.plot(google_data_pre_2014['date'], google_data_pre_2014['open'], 
         color = 'k', label = 'NYSE data')


plt.xlabel('date')
plt.xticks(rotation=45)
plt.ylabel('opening price (\$)')

skip = 100
#^^^this variable sets the spacing of the error bars, smaller=closer together
#^^^larger=farther apart
plt.errorbar(google_data_pre_2014['date'][::skip], price_model_avg[::skip],
             yerr = price_model_std[::skip], 
             fmt = 'o',
             c = 'r', 
             label = 'model result', 
             zorder = 3)

plt.legend()
plt.show()   
````

````python
plt.plot(google_data_pre_2014['date'], price_model_avg)
plt.xticks(rotation=45)
plt.xlabel('date')
plt.ylabel('average price ($)')
plt.show()
````

## Wrapping up

In this analysis, I went through data exploration, analysis, and Monte Carlo model prediction. The average random walk should resemble a straight line. There are further insights you can get by analyzing the random walk data, but for now it looks like we can accurately predict the growth of GOOGL stock over four years. __What are some caveats to this method? If we continue to predict prices into 2015, what would happen compared to the real data?__

Analyzing the stock market can be very difficult because there are many factors that go into how well a particular company does. Take, this GOOGL data for example, we only analyzed the data before a significant drop in its value. This drop was due to multiple factors including concerns about ad pricing and mobile revenue and the company's heavy spending. There was also a stock split that Google did where they split up into two classes of shares: Class A 'GOOGL' and Class B 'GOOG'. An analysis like what we did does not account for factors like these. If we were to find a best fit line for the average data and use it to predict prices into 2015, the large drop would not be included and the price of 'GOOGL' would be much higher than what the actual price is now.

````python
nvidia_data = data[data['symbol'] == 'NVDA']
nvidia_date = pd.to_datetime(nvidia_data['date'])

plt.plot(nvidia_data['date'], nvidia_data['open'])
plt.xlabel('date')
plt.ylabel('opening price (\$)')
plt.show()
````

````python
nvidia_dprice = np.diff(nvidia_data['open'])  
#^^^np.diff finds the difference between each day and the next

plt.plot(nvidia_data['date'][1:], nvidia_dprice)
plt.xlabel('date')
plt.ylabel('change in opening price (\$/day)')
plt.show()
````

````python
nvidia_mean_dprice = np.mean(nvidia_dprice)
print(nvidia_mean_dprice)

nvidia_std_dprice = np.std(nvidia_dprice)
print(nvidia_std_dprice)

x1 = np.linspace(-5, 5)
from scipy import stats
nvidia_price_pdf = stats.norm.pdf(x1, loc = nvidia_mean_dprice, 
                                  scale = nvidia_std_dprice)
#^^^loc = mean of the data, scale = std of the data
#^^^norm.pdf sets the data as a probability density function
````

````python
plt.hist(nvidia_dprice, bins=100, density=True)
#^^^plots the price difference as a histogram
#^^^when density=True, returns a probability density so that the area under
#^^^integrates to 1 centered at 0
plt.plot(x1, nvidia_price_pdf)
plt.title('NVDA changes in price\n'+
         'avg: \${:.2f} stdev: \${:.2f}'.format(nvidia_mean_dprice, 
                                                nvidia_std_dprice))
plt.show()
````

````python
rng = default_rng(42)
#^^^fixes the random number seed

N_models = 100
nvidia_dprice_model = rng.normal(size = (len(nvidia_data), N_models), 
                          loc = 0.053, scale = 0.791)
#^^^loc=mean_dprice; scale=std_dprice

print(len(nvidia_dprice_model[0,:]))
#^^^dprice_model has different 100 random models - 100 columns - for each day

print(len(nvidia_dprice_model[:,0]))
#^^^dprice_model has 1762 simulated days - 1762 rows - for each model
#^^^matches the length of nvidia_data

plt.hist(nvidia_dprice, bins=100, density=True, label = 'NYSE data')
plt.plot(x1, nvidia_price_pdf)
plt.hist(nvidia_dprice_model[:, 0], 100, density = True, 
         histtype = 'step', linewidth = 3, 
         label = 'model prediction 1')
#^^^histtype genrates a line plot unfilled default is 'bar'

plt.title('NVDA changes in price\n'+
         'avg: \${:.2f} stdev: \${:.2f}'.format(nvidia_mean_dprice, 
                                                nvidia_std_dprice))
plt.legend()
plt.show()
````

````python
nvidia_price_model = np.cumsum(nvidia_dprice_model, axis = 0) + nvidia_data['open'].values[0]

plt.plot(nvidia_data['date'], nvidia_price_model, alpha = 0.3)
plt.plot(nvidia_data['date'], nvidia_data['open'], c = 'k',
         label = 'NVDA data')

plt.xlabel('date')
plt.ylabel('opening price (\$)')
plt.legend()
plt.show()
````

````python
nvidia_price_model_avg = np.mean(nvidia_price_model, axis = 1)
nvidia_price_model_std = np.std(nvidia_price_model, axis = 1)
#^^^get the mean and std across the 100 columns ie for each model simulation
#^^^the mean and std is recorded for each day- length of nvidia_data
print(len(nvidia_price_model_avg))

plt.plot(nvidia_data['date'], nvidia_price_model, alpha = 0.3)
plt.plot(nvidia_data['date'], nvidia_data['open'], 
         color = 'k', label = 'NVDA data')

plt.xlabel('date')
plt.ylabel('opening price (\$)')

skip = 100
#^^^this variable sets the spacing of the error bars, smaller=closer together
#^^^larger=farther apart
plt.errorbar(nvidia_data['date'][::skip], nvidia_price_model_avg[::skip],
             yerr = nvidia_price_model_std[::skip], 
             fmt = 'o',
             c = 'r', 
             label = 'model result', 
             zorder = 3)

plt.legend()
plt.show()  
````

````python
a_1n, a_0n = np.polyfit(range(len(nvidia_data['date'])), 
                        nvidia_price_model_avg, deg=1)
print(a_1n, a_0n)
f_linear = lambda x: a_1n*x + a_0n


plt.plot(nvidia_data['date'], nvidia_price_model_avg, label='NVDA avg price')
plt.plot(nvidia_data['date'], f_linear(range(len(nvidia_data['date']))), 
         'k--', linewidth=2, label='best fit')
plt.xlabel('date')
plt.ylabel('average price ($)')
plt.title("NVDA Average Opening Price")
plt.legend()
plt.show()
````
